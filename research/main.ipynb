```python
import os
import re
import json
import argparse
from pathlib import Path
from urllib.parse import urlparse

import requests
from dotenv import load_dotenv
```

```python
ACTOR_URL = "https://api.apify.com/v2/acts/apify~instagram-profile-scraper/run-sync-get-dataset-items"


def get_username(instagram_url: str) -> str:
    instagram_url = instagram_url.strip()

    if instagram_url.startswith("@"):
        return instagram_url[1:]

    if "instagram.com" not in instagram_url and "/" not in instagram_url:
        return instagram_url

    if not instagram_url.startswith(("http://", "https://")):
        instagram_url = "https://" + instagram_url

    parsed = urlparse(instagram_url)
    parts = [p for p in parsed.path.split("/") if p]

    if not parts:
        raise ValueError("Не смог достать username из ссылки")

    username = parts[0]

    if username in {"p", "reel", "stories", "explore", "accounts"}:
        raise ValueError("Нужна ссылка на профиль, а не на пост/reel/story")

    return username


def safe_name(value: str) -> str:
    value = str(value or "unknown")
    return re.sub(r"[^\w.-]+", "_", value).strip("_")


def run_apify(username: str) -> dict:
    if not APIFY_TOKEN:
        raise RuntimeError("Нет APIFY_TOKEN в .env")

    payload = {
        "usernames": [username],
        "includeAboutSection": False
    }

    response = requests.post(
        ACTOR_URL,
        params={
            "format": "json",
            "clean": "true"
        },
        headers={
            "Authorization": f"Bearer {APIFY_TOKEN}",
            "Content-Type": "application/json"
        },
        json=payload,
        timeout=300
    )

    response.raise_for_status()

    items = response.json()

    if not items:
        raise RuntimeError(f"Apify ничего не вернул по @{username}")

    return items[0]


def is_video(post: dict) -> bool:
    post_type = str(post.get("type", "")).lower()
    product_type = str(post.get("productType", "")).lower()

    return (
        "video" in post_type
        or "reel" in post_type
        or "clips" in product_type
        or bool(post.get("videoUrl"))
        or bool(post.get("videoPlayCount"))
        or bool(post.get("videoViewCount"))
    )


def download_file(url: str, path: Path) -> bool:
    if not url:
        return False

    headers = {
        "User-Agent": "Mozilla/5.0",
        "Referer": "https://www.instagram.com/"
    }

    try:
        with requests.get(url, headers=headers, stream=True, timeout=120) as r:
            r.raise_for_status()

            path.parent.mkdir(parents=True, exist_ok=True)

            with open(path, "wb") as f:
                for chunk in r.iter_content(chunk_size=1024 * 1024):
                    if chunk:
                        f.write(chunk)

        return path.exists() and path.stat().st_size > 0

    except Exception as e:
        print(f"Не скачалось: {url[:80]}...")
        print(f"Ошибка: {e}")
        return False


def normalize_profile(profile: dict, downloaded_videos: list) -> dict:
    return {
        "id": profile.get("id"),
        "username": profile.get("username"),
        "full_name": profile.get("fullName"),
        "followers": profile.get("followersCount"),
        "following": profile.get("followsCount"),
        "posts_count": profile.get("postsCount"),
        "bio": profile.get("biography"),
        "external_url": profile.get("externalUrl"),
        "profile_url": profile.get("url"),
        "profile_pic_url": profile.get("profilePicUrl"),
        "verified": profile.get("verified"),
        "private": profile.get("private"),
        "business_account": profile.get("isBusinessAccount"),
        "business_category": profile.get("businessCategoryName"),
        "videos": downloaded_videos
    }


def save_profile(instagram_url: str, output_root: str = "output") -> Path:
    username = get_username(instagram_url)
    print(f"Парсю @{username}")

    profile = run_apify(username)

    profile_id = profile.get("id") or profile.get("username") or username
    profile_dir = Path(output_root) / safe_name(profile_id)
    videos_dir = profile_dir / "videos"

    profile_dir.mkdir(parents=True, exist_ok=True)
    videos_dir.mkdir(parents=True, exist_ok=True)

    latest_posts = profile.get("latestPosts") or []

    video_posts = [post for post in latest_posts if is_video(post)][:20]

    videos_info = []

    for i, post in enumerate(video_posts, start=1):
        shortcode = post.get("shortCode") or post.get("shortcode") or post.get("id") or f"video_{i}"
        video_url = post.get("videoUrl")

        local_path = None
        ok = False

        # Скачиваем только первые 3 видео
        if i <= 3:
            filename = f"{i:02d}_{safe_name(shortcode)}.mp4"
            local_path = videos_dir / filename
            ok = download_file(video_url, local_path)

        item = {
            "number": i,
            "id": post.get("id"),
            "shortcode": shortcode,
            "post_url": post.get("url") or f"https://www.instagram.com/p/{shortcode}/",
            "caption": post.get("caption"),
            "timestamp": post.get("timestamp"),
            "likes": post.get("likesCount"),
            "comments": post.get("commentsCount"),
            "views": post.get("videoViewCount") or post.get("videoPlayCount") or post.get("viewsCount"),
            "thumbnail": post.get("displayUrl") or post.get("thumbnailUrl"),
            "video_url": video_url,
            "local_path": str(local_path) if ok else None,
            "downloaded": ok
        }

        videos_info.append(item)

        if i <= 3:
            if ok:
                print(f"Скачано: {local_path}")
            else:
                print(f"Не скачалось видео {shortcode}")

    normalized = normalize_profile(profile, videos_info)

    with open(profile_dir / "profile.json", "w", encoding="utf-8") as f:
        json.dump(normalized, f, ensure_ascii=False, indent=2)

    with open(profile_dir / "bio.txt", "w", encoding="utf-8") as f:
        f.write(f"ID: {normalized['id']}\n")
        f.write(f"Username: {normalized['username']}\n")
        f.write(f"Full name: {normalized['full_name']}\n")
        f.write(f"Followers: {normalized['followers']}\n")
        f.write(f"Following: {normalized['following']}\n")
        f.write(f"Posts: {normalized['posts_count']}\n")
        f.write(f"External URL: {normalized['external_url']}\n\n")
        f.write(normalized["bio"] or "")

    print()
    print(f"Готово. Папка: {profile_dir}")

    return profile_dir
```
